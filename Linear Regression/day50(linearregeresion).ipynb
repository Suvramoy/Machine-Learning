```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#input column
x=df.iloc[:,0:1]
#output coloumn
y=df.iloc[:,1:]   #or[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)#0.2 means 20 percent data give to the test
```

READY TO APPLY THE ALGO

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()   # making objexct
```

```python
#for train the data using fit
lr.fit(x_train,y_train)
```

```python
x_test
```

```python
y_test
```

```python
lr.predict(x_test.iloc[0]) #this code give us errror 
#beacuse we train our algo the input shape is (1,1)1 row and 1 coloumn but we give 1D data
```

```python
lr.predict(x_test.iloc[0].values.reshape(1,1))
```

```python
lr.predict(x_test.iloc[2].values.reshape(1,1))
```

```python
plt.scatter(df['cgpa'],df['package']);
plt.xlabel('CGPA');
plt.plot(x_test,lr.predict(x_test),color='red');#we dont give df[cgpa because we only train our model for 20%data i.e 40 data out of 200 data]
plt.ylabel('PACKAGE(in lpa)');
```

```python
plt.scatter(df['cgpa'],df['package']);
plt.xlabel('CGPA');
plt.plot(x_train,lr.predict(x_train),color='red');#whole data
plt.ylabel('PACKAGE(in lpa)');
```
